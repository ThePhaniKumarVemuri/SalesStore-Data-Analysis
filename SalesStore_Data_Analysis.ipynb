```python
df.dropna(inplace=True)  # Drop missing values if necessary
```

```python
# Convert Order Date to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.strftime("%Y-%m")  # Extract Year-Month
df.describe()
```

```python
# 5. Find duplicates
print("\nNumber of Duplicate Rows:", df.duplicated().sum())
```

```python
# 6. Check correlation between numerical columns
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))
```

```python
# 7. Count values for categorical columns (to see distribution)
print("\nCategory Value Counts:\n", df["Category"].value_counts())
print("\nRegion Value Counts:\n", df["Region"].value_counts())
```

```python
# 8. Get the most and least profitable product
most_profitable = df.loc[df["Profit"].idxmax()]
least_profitable = df.loc[df["Profit"].idxmin()]
print("\nMost Profitable Product:\n", most_profitable)
print("\nLeast Profitable Product:\n", least_profitable)
```

```python
# 1. Sales Distribution by Region
sales_by_region = df.groupby("Region")["Sales"].sum().reset_index()
print("\nSales Distribution by Region: \n")
print(sales_by_region)
```

```python
# 2. Category-Wise Sales & Profit
category_sales_profit = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()
print("\nCategory-wise Sales & Profit: \n")
print(category_sales_profit)
```

```python
# 3. Top 10 Cities by Sales
top_cities = df.groupby("City")["Sales"].sum().nlargest(10).reset_index()
print("\nTop 10 Cities by Sales: \n")
print(top_cities)
```

```python
# 4. Monthly Sales Trend
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
print("\nMonthly Sales Trend: \n")
print(monthly_sales)
```

```python
# 5. Discount Impact on Profitability
discount_impact = df.groupby("Discount")[["Sales", "Profit"]].mean().reset_index()
print("\nDiscount Impact on Profitability: \n")
print(discount_impact)
```

```python
# 6. Top-Selling Sub-Categories
top_subcategories = df.groupby("Sub-Category")["Sales"].sum().nlargest(10).reset_index()
print("\nTop-Selling Sub-Categories:")
print(top_subcategories)
```

```python
# Convert Order Date to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.strftime("%Y-%m")  # Extract Year-Month
```

```python
# 1. Sales Distribution Across Regions (Bar Chart)
region_sales = df.groupby("Region")["Sales"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=region_sales, x="Region", y="Sales", palette="Set2")
plt.title("Sales Distribution Across Regions")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
# 2. Category-Wise Sales & Profit (Grouped Bar Chart)
cat_data = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()
x = np.arange(len(cat_data["Category"]))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, cat_data["Sales"], width, label="Sales")
plt.bar(x + width/2, cat_data["Profit"], width, label="Profit")
plt.xticks(x, cat_data["Category"])
plt.title("Sales & Profit by Category")
plt.ylabel("Amount")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 3. Top 10 Cities by Sales (Horizontal Bar Chart)
top_cities = df.groupby("City")["Sales"].sum().nlargest(10).sort_values().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(data=top_cities, y="City", x="Sales", hue="City", palette="viridis", dodge=False, legend=False)
plt.title("Top 10 Cities by Sales")
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
# 4. Sales vs. Profit Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Sales", y="Profit", hue="Category", size=np.abs(df["Profit"]),
                alpha=0.6, palette="deep")
plt.title("Sales vs. Profit Scatter Plot")
plt.tight_layout()
plt.show()
```

```python
# 5. Monthly Sales Trend (Line Chart)
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_sales, x="Month", y="Sales", marker='o')
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```

```python
# 6. Discount Impact on Profitability (Scatter Plot)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Discount", y="Profit", hue="Category", size=np.abs(df["Profit"]),
                alpha=0.5, palette="Set1")
plt.title("Discount Impact on Profit")
plt.tight_layout()
plt.show()
```

```python
# 7. Top-Selling Sub-Categories (Pie Chart)
top_subcategories = df.groupby("Sub-Category")["Sales"].sum().nlargest(10).reset_index()
plt.figure(figsize=(8,8))
plt.pie(top_subcategories["Sales"], labels=top_subcategories["Sub-Category"],
        autopct="%1.1f%%", startangle=140, wedgeprops=dict(width=0.4))
plt.title("Top-Selling Sub-Categories")
plt.tight_layout()
plt.show()
```

```python
# 8. Heatmap of Sales and Profit Across Regions & Categories
heatmap_data = df.groupby(["Category", "Region"])["Sales"].sum().unstack()
plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap of Sales Across Regions & Categories")
plt.tight_layout()
plt.show()
```
